```python
def cos_sim(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return dot_product / (norm_a * norm_b)
```

```python
matrix = []
for i in range(1000):
    a = data.loc[:,i]
#     print(a)
    cols = []
    for j in range(1000):
        b = data.loc[:,j]
#         print(b)
        similarity = cos_sim(a, b)
        cols.append(similarity)
    matrix.append(cols)
```

```python
def predict(userId, movieId):
    dict_with_rating = {}
    lis = []

    for idx, rating in enumerate(df_movie_features.iloc[userId]):
        if rating == 'X':
            continue
        if(rating > 0):
            temp = (idx, rating)
            dict_with_rating[idx] = rating
            lis.append(temp)
    top_k = []
    for i, sim in enumerate(utility_matrix.iloc[movieId]):
        for j, k in lis:
            if(j == i):
                top_k.append([i,sim])
            
    
    top_sorted_sim = sorted(top_k, key=lambda l: l[1], reverse=True)
    top_5 = top_sorted_sim[:5]
    sum_ = 0
    sim_sum = 0
    for top in top_5:
        sum_ += dict_with_rating[top[0]]*top[1]
        sim_sum += top[1]
    return sum_ // sim_sum

predict(14, 1)
```
